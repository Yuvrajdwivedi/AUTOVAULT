```python
from ultralytics import YOLO
import os
import shutil

# Path where kagglehub downloaded the dataset (using the 'path' variable from kernel state)
kaggle_download_path = path

# Define the base directory for the dataset as specified in dataset.yaml
base_dataset_path = "/content/pothole_ai"
# Define the target directory for the actual dataset content as derived from dataset.yaml
target_dataset_content_path = os.path.join(base_dataset_path, "datasets")

# Create the target directory if it doesn't exist
os.makedirs(target_dataset_content_path, exist_ok=True)

# Copy contents from the downloaded path to the expected structure
# Assuming the downloaded content has 'train', 'val', 'test' directories directly inside the kaggle_download_path
print(f"Moving dataset from {kaggle_download_path} to {target_dataset_content_path}...")
for item in os.listdir(kaggle_download_path):
    src_item_path = os.path.join(kaggle_download_path, item)
    dest_item_path = os.path.join(target_dataset_content_path, item)
    if os.path.isdir(src_item_path):
        # Use shutil.copytree for directories, ensuring intermediate directories are created
        shutil.copytree(src_item_path, dest_item_path, dirs_exist_ok=True)
    else:
        # Use shutil.copy2 for files (preserves metadata)
        shutil.copy2(src_item_path, dest_item_path)
print("Dataset movement complete.")

model = YOLO("yolov8n.pt")

# The dataset_yaml_path variable is available from previous cell execution
# It contains '/content/pothole_ai/dataset.yaml'
model.train(
    data=dataset_yaml_path,
    epochs=50,
    imgsz=640,
    batch=16
)
```

```python
import numpy as np

class KalmanFilter:
    def __init__(self):
        self.x = np.array([[0], [0]])  # position + velocity
        self.P = np.eye(2) * 1000
        self.F = np.array([[1, 1], [0, 1]])
        self.H = np.array([[1, 0]])
        self.R = np.array([[5]])
        self.Q = np.eye(2)

    def predict(self):
        self.x = self.F @ self.x
        self.P = self.F @ self.P @ self.F.T + self.Q
        return self.x

    def update(self, z):
        y = z - (self.H @ self.x)
        S = self.H @ self.P @ self.H.T + self.R
        K = self.P @ self.H.T @ np.linalg.inv(S)
        self.x = self.x + K @ y
        self.P = (np.eye(2) - K @ self.H) @ self.P
        return self.x

kf = KalmanFilter()

ultrasonic_reading = np.array([[120]])  # cm distance
kf.predict()
estimate = kf.update(ultrasonic_reading)
print("Estimated Distance:", estimate[0][0])
```

```python
def fuzzy_brake(distance, speed, lean_angle):
    # distance in meters
    # speed in km/h
    # lean_angle in degrees

    risk = 0

    # Distance contribution
    if distance < 2:
        risk += 0.5
    elif distance < 5:
        risk += 0.3
    else:
        risk += 0.1

    # Speed contribution
    if speed > 50:
        risk += 0.4
    elif speed > 20:
        risk += 0.2
    else:
        risk += 0.1

    # Lean angle safety penalty
    if lean_angle > 25:
        risk -= 0.3   # reduce braking during sharp turn

    # Clamp risk between 0 and 1
    risk = max(0, min(1, risk))

    # Convert risk to brake intensity
    if risk < 0.3:
        return "LOW ALERT", 0.0
    elif risk < 0.6:
        return "MEDIUM ALERT", 0.3
    elif risk < 0.8:
        return "HIGH BRAKE", 0.6
    else:
        return "EMERGENCY BRAKE", 1.0

state, brake_force = fuzzy_brake(distance=1.5, speed=60, lean_angle=10)
print(state, brake_force)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

X = np.random.rand(1000, 20, 3)   # 20 timesteps, 3 features (speed, accel, jerk)
y = np.random.randint(0, 2, 1000) # 0=normal road, 1=traffic mode

model = Sequential([
    LSTM(64, input_shape=(20,3)),
    Dense(32, activation="relu"),
    Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model.fit(X, y, epochs=10, batch_size=32)
```

```python
pred = model.predict(np.random.rand(1,20,3))
print("Traffic Mode Probability:", pred[0][0])
```

```python
def autovault_decision(yolo_detected, distance, speed, lean_angle, traffic_prob):

    if traffic_prob > 0.7 and speed < 10:
        return "TRAFFIC MODE ACTIVE (NO BRAKING)", 0.0

    if not yolo_detected and distance > 5:
        return "NO THREAT", 0.0

    state, brake_force = fuzzy_brake(distance, speed, lean_angle)

    if yolo_detected:
        return state, brake_force
    else:
        return "SENSOR ALERT ONLY", 0.1

decision, brake = autovault_decision(
    yolo_detected=True,
    distance=1.8,
    speed=55,
    lean_angle=5,
    traffic_prob=0.2
)

print(decision, brake)
```
